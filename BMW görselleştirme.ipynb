```python
annual_avg = dt.groupby(dt['Date'].dt.year)['Close'].mean()
```

```python
print(f"Yıllık ortalama:{annual_avg}")
```

```python
plt.figure(figsize=(10,6))
annual_avg.plot(kind='line', marker='o', color='blue', linestyle='-')
plt.title('Yıllara göre ortalama satış')
plt.xlabel('Yıl')
plt.ylabel('Ort')
plt.grid(True)
plt.show()
```

```python
annual_avg_low = dt.groupby(dt['Date'].dt.year)['Low'].mean()
annual_avg_high = dt.groupby(dt['Date'].dt.year)['High'].mean()
```

```python
plt.figure(figsize=(20,10))
annual_avg_high.plot(kind='line', marker='o', color='red', label='Ortalama en yüksek')
annual_avg_low.plot(kind='line', marker='o', color='blue', label='Ortalama en düşük')
plt.title('Yıllara göre ortalama en yüksek - en düşük değerler arasındaki karşılaştırma')
plt.xlabel('Yıl')
plt.ylabel('Ort')
plt.grid(True)
plt.legend()
plt.show()
```

```python
dt_yeni = dt.copy()
dt_yeni.set_index('Date', inplace=True)
dt_yeni.info()
```

```python
dt_yeni
```

```python
dt_aralik = dt_yeni.loc['2020-01-01':'2024-12-30']
```

```python
rolling_mean_short = dt_aralik['Close'].rolling(window=20).mean()
rolling_mean_long = dt_aralik['Close'].rolling(window=100).mean()

plt.figure(figsize=(15, 7))

plt.plot(dt_aralik.index, dt_aralik['Close'], color='blue', linewidth=1, alpha=0.7, label='Kapanış Fiyatı')
plt.plot(dt_aralik.index, rolling_mean_short, color='orange', linewidth=2, label='20 Günlük Ortalama')
plt.plot(dt_aralik.index, rolling_mean_long, color='red', linewidth=2, label='100 Günlük Ortalama')

plt.title('Kapanış Fiyatı ve Hareketli Ortalamalar (2020-2024)', fontsize=18)
plt.xlabel('Tarih', fontsize=14)
plt.ylabel('Fiyat', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.tight_layout()

plt.show()
```
